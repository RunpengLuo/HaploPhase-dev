```python
fig = plt.figure()
fig, axes =plt.subplots(2,2)
fig.set_figheight(6)
fig.set_figwidth(6)

# labels=['HiC-Pro', 'Kmer2SNP', 'Phasing Algorithm']
labels=['K', 'H', 'P']
pt76_verkko = [0.52, 21.32, 0.03]
pt76_hifiasm = [0.52, 19.16,  0.04]

pst_verkko = [0.13, 14.52, 0.01]
pst_hifiasm = [0.13, 13.75, 0.01]

colors = ['yellowgreen','lightskyblue','red']

def my_autopct(pct):
    return ('%.2f' % pct) if pct < 20 else ''

def perc_pie(x, labels, ax, title):
    s = sum(x)
    percx = [round(v*100/s, 2) for v in x]
    label_x = [f"{v}%" for (l,v) in zip(labels, percx)]
    ax.pie(percx, explode=(0,0.2,0.1), colors=colors, labels=label_x, autopct=my_autopct)
    ax.set_title(title)


perc_pie(pt76_verkko, labels, axes[0][0], "(a)(i)")
perc_pie(pt76_hifiasm, labels, axes[0][1], "(a)(ii)")
perc_pie(pst_verkko, labels, axes[1][0], "(b)(i)")
perc_pie(pst_hifiasm, labels, axes[1][1], "(b)(ii)")

plt.tight_layout()

```
